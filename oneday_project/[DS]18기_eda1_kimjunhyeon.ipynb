````python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import Chrome
from bs4 import BeautifulSoup
driver = webdriver.Chrome(executable_path='../driver/chromedriver.exe')
driver.get("https://www.starbucks.co.kr/")
````

````python
import warnings
warnings.simplefilter(action='ignore')
````

````python
# 스토어 탭 클릭
store = driver.find_element(By.CSS_SELECTOR, "#gnb > div > nav > div > ul > li.gnb_nav03 > h2 > a")
store.click()
````

````python
# 매장 찾기
search_store = driver.find_element(By.CSS_SELECTOR, "#container > section.store_bn1_section > article > div.store_bn1_btn > a")
search_store.click()
````

````python
# 지역 검색
search_local = driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a")
search_local.click()
````

````python
# 도시 선택
select_city = driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a")
select_city.click()
````

````python
# 전체 탭 클릭
driver.find_element(By.CSS_SELECTOR, "#mCSB_2_container > ul > li:nth-child(1) > a").click()
````

````python
# 서울내 전체 리스트 가져오기
# li_list = driver.find_elements(By.CLASS_NAME, "mCSB_container ul li") # li에 담긴 정보들 가져오기. 직접 접근 하려했던것이 실수.
# len(li_list) # 담긴 갯수확인.
'''
<div id="mCSB_3_container" class="mCSB_container" style="position: relative; top: 0px; left: 0px;" dir="ltr">
    <ul>
        <li>
            <strong> 이름 </strong>
            <p> 주소 </p>
            
        </li>
    </ul>
</div>


'''
````

````python
#현재 페이지 html가져오기
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
````

````python
# html에서 클래스네임 이용하여 li 영역 가져오기
seoulList = soup.select("#mCSB_3_container > ul > li")
len(seoulList), seoulList
````

````python
# 하나 꺼내서 확인해보기
seoulList[0]['data-name'],  seoulList[0].select_one('.result_details').text, seoulList[0]['data-lat'], seoulList[0]['data-long']
# 'data-name' 속성으로도 부분 추출 가능한점 확인하기
# li 속 또다른 p태그에 있는 클래스 속성값을 유의하여 select 이용하여 주소 가져오기.
# select도 리스트 형태로 반환? select_one를 써서 단일값으로 가져와야 .text 를 가져올 수 있음.
````

````python
seoulList[0].select_one('.result_details').text[:-9] # 형식이 같아서 뒤에서 9개는 다 날려도됨을 확인.
#seoulList[0].select_one('p').text[:-9] - 같은내용
````

````python
title = seoulList[0].select_one("strong").text[:-2] # 지점 제외
lat = seoulList[0].get('data-lat')
lng = seoulList[0].get('data-long')
address = seoulList[0].select_one('.result_details').text[:-9]
gu = seoulList[0].select_one('.result_details').text.split(" ")[1]
title, lat, lng, address, gu
````

````python
from tqdm import notebook
import time
````

````python
data = []
for idx in notebook.tqdm(seoulList):
    title = idx.select_one("strong").text[:-2],
    address = idx.select_one('.result_details').text[:-9],
    lat = idx.get('data-lat'),
    lng = idx.get('data-long')
    gu = idx.select_one('.result_details').text.split(" ")[1]

    data.append({
        "title" : title,
        "address" : address,
        "lat" : lat,
        "lng" : lng,
        "gu" : gu
    })
data
driver.close()
````

````python
df = pd.DataFrame(data)
df
````

````python
df.to_excel("../data/starbucks_seoulList.xlsx", encoding="utf-8")
````
