```python
# 1-3
def get_alpha2(country_name):
    try:
        country = pycountry.countries.get(name=country_name)
        if country:
            return country.alpha_2
        else:
            return None
    except Exception as e:
        return None

df_target['code'] = df_target['country'].apply(get_alpha2)
df_target['code'].isnull().sum()
```

```python
null_rows = df_target[df_target['code'].isnull()]
null_rows
```
