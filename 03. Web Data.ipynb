````python
import requests # 거의비슷= from urllib.request.Request
from bs4 import BeautifulSoup
````

````python
url = "https://finance.naver.com/marketindex/"
response = requests.get(url)
# requests.get(), requests.post() - 두가지 방식 존재.
# response.text    --- response에 이미 담겨있음을 확인가능
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
````

````python
# list 데이터 긁어오기
#선택자 class - . / id - # / 하위 - >
#soup.find_all("li", "on")
exchangeList = soup.select('#exchangeList > li')
len(exchangeList) , exchangeList
````

````python
# select method
title = exchangeList[0].select_one('.h_lst').text
exchange = exchangeList[0].select_one('.value').text
change = exchangeList[0].select_one('.change').text
updown = exchangeList[0].select_one('.head_info.point_dn > .blind').text #띄어쓰기 유의 선택자결정시 '공백'과 '>' 차이 알아둘 것

title, exchange, change, updown
````

````python
baseUrl = "https://finance.naver.com"
baseUrl + exchangeList[0].select_one('a').get('href')
````

````python
exchangeList.select_one('h3')
````

````python
# for item in exchangeList:
#     data = {
#         "title": item.select_one('.h_lst').text,
#         "exchange": item.select_one(".value").text,
#         "change": item.select_one('.change').text,
#         "updown": item.select_one('.blind').text
#     }
#     print(data)
````

````python
exchange_datas = []
baseUrl = "https://finance.naver.com"

for item in exchangeList:
    # 시도해보는 클래스 리스트
    classes_to_try = ['head_info.point_dn', 'head_info.point_up']

    # 기본 updown 값을 설정
    updown = None

    # 각 클래스에 대해 시도
    for class_name in classes_to_try:
        element = item.select_one(f'.{class_name} > .blind')
        if element is not None:
            updown = element.text
            break

    data = {
        "title": item.select_one('.h_lst').text,
        "exchange": item.select_one(".value").text,
        "change": item.select_one('.change').text,
        "updown": updown,
        "link" : baseUrl + item.select_one('a').get('href')
    }
    exchange_datas.append(data)

for data in exchange_datas:
    print(data)
````

## 3. 시카고 맛집 데이터 분석 - 메인페이지

````python
# !pip install fake-useragent
from urllib.request import Request, urlopen
from fake_useragent import UserAgent
from bs4 import BeautifulSoup

url_base = "https://www.chicagomag.com/"
url_sub = "chicago-magazine/november-2012/best-sandwiches-chicago/"
url = url_base + url_sub
#url

ua = UserAgent()
ua.ie

# headers - request 접근에 문제가 생길 때는 헤더값을 준다.
# req = Request(url, headers={"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}) 
# req = Request(url, headers={"User-Agent" : "Chrome"})
req = Request(url, headers={"User-Agent" : ua.ie})
html = urlopen(req)
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
````
